```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
probabilities = [0.983, 0.017]  # 98.3% for ln(0.65), 1.7% for 0
values = [0, np.log(0.65)]  # Corresponding values

# Generate a list of random numbers with the given probabilities
v = np.random.choice(values, size=10000, p=probabilities)

v
```

```python
len(v)
```

```python
df = pd.DataFrame({'v': v})
df[df['v']==0]
```

```python
epsilon = np.random.normal(0,1,10000)

epsilon
```

```python
g = np.exp(0.02 + 0.02*epsilon + v)
```

```python
g
```

```python
df['g'] = g
df['epsilon'] = epsilon
```

```python
df
```

```python
gamma = np.arange(1,4.1,0.1)
```

```python
gamma
```

```python
for i in gamma:
    i = round(i,1)
    df['M'+str(i)] = 0.99* g**(-i)
```

```python
df
```

```python
mean_gamma_list = []
std_gamma_list = []
for i in gamma:
    i = round(i,1)
    mean_gamma = df['M'+str(i)].mean()
    std_gamma = df['M'+str(i)].std()
    mean_gamma_list.append(mean_gamma)
    std_gamma_list.append(std_gamma)
```

```python
mean_gamma_arr = np.array(mean_gamma_list)
mean_gamma_arr
```

```python
std_gamma_arr = np.array(std_gamma_list)
std_gamma_arr
```

```python
volatility_ratio = std_gamma_arr / mean_gamma_arr
```

```python
volatility_ratio
```

```python
df2 = pd.DataFrame({'gamma': gamma, 'volatility_ratio' : volatility_ratio, 'mean_gamma':mean_gamma_arr, 'std_gamma' : std_gamma_arr}).set_index('gamma')
```

```python
df2
```

```python
min_gamma = df2[df2['volatility_ratio']>0.4].index.min()
min_gamma
```

```python
min_vol_ratio = df2['volatility_ratio'][df2['volatility_ratio']>0.4].min()
min_vol_ratio
```

```python
plt.figure(figsize = [12,7])
plt.scatter(gamma, volatility_ratio)
plt.plot(gamma, volatility_ratio,color = 'black')
# plt.scatter(min_gamma, min_vol_ratio , color = 'red')
plt.xlabel('Gamma')
plt.ylabel('Volatility Ratio')
plt.title ('Volatility Ratio vs Gamma')
plt.xlim(1,4.0)
plt.ylim(0,0.7)
plt.xticks(np.arange(1 ,4.1, 0.1))
# plt.axvline(min_gamma, ymin = 0, ymax = 4/7, linestyle = '-.', color = 'red')
# plt.axhline(min_vol_ratio, xmin = 0, xmax = 2.3/3, linestyle = '-.', color = 'red')
plt.legend()
```

```python
plt.figure(figsize = [12,7])
plt.scatter(gamma, volatility_ratio)
plt.plot(gamma, volatility_ratio,color = 'black')
plt.scatter(min_gamma, min_vol_ratio , color = 'red')
plt.xlabel('Gamma')
plt.ylabel('Volatility Ratio')
plt.title ('Volatility Ratio vs Gamma')
plt.xlim(1,4.0)
plt.ylim(0,0.7)
plt.xticks(np.arange(1 ,4.1, 0.1))
plt.axvline(min_gamma, ymin = 0, ymax = 4/7, linestyle = '-.', color = 'red')
plt.axhline(min_vol_ratio, xmin = 0, xmax = 2.3/3, linestyle = '-.', color = 'red')
plt.legend()
```
